```python
import pandas as pd

# Load dataset, raw data sourced from Our World in Data and World Bank
df = pd.read_csv("if you were born here?/life_expectancy.csv")


```

```python
# List of selected countries
countries = [
    "United States", "Germany", "Japan", "Canada", "Australia",
    "China", "Brazil", "Mexico", "South Africa", "Turkey",
    "India", "Vietnam", "Zimbabwe", "Egypt", "Bangladesh",
    "Ethiopia", "Afghanistan", "Niger", "Chad", "Somalia"
]


df.columns = df.columns.str.strip()

# Filter dataset
filtered = df[
    (df["Entity"].isin(countries)) &
    (df["Year"].between(2000, 2023))
]

# Group by country and compute average of available years
result = (
    filtered
    .groupby("Entity")["Life expectancy"]
    .mean()
    .reset_index()
)

# Rename columns
result.columns = ["Country", "Life expectancy"]


result["Life expectancy"] = result["Life expectancy"].round(2)

print(result)

```
