```python
import pandas as pd

# Load dataset (skip first 3 empty rows)
df = pd.read_csv("gni_ppp.csv", skiprows=3)

# Countries we want
countries = [
    "United States", "Germany", "Japan", "Canada", "Australia",
    "China", "Brazil", "Mexico", "South Africa", "Turkiye",
    "India", "Vietnam", "Zimbabwe", "Egypt, Arab Rep.", "Bangladesh",
    "Ethiopia", "Afghanistan", "Niger", "Chad", "Somalia, Fed. Rep., Viet Nam"
]
```

```python
# Keep only these countries
df = df[df["Country Name"].isin(countries)].copy()

# Convert year columns to numeric (important for averaging)
years = [str(y) for y in range(2000, 2024)]
df[years] = df[years].apply(pd.to_numeric, errors="coerce")

# Compute averages
df["GNI 1"] = df[[str(y) for y in range(2000, 2006)]].mean(axis=1)
df["GNI 2"] = df[[str(y) for y in range(2006, 2011)]].mean(axis=1)
df["GNI 3"] = df[[str(y) for y in range(2000, 2024)]].mean(axis=1)

# Keep only required columns
final_df = df[["Country Name", "GNI 1", "GNI 2", "GNI 3"]]

# Optional rounding
final_df = final_df.round(2)

# Save
final_df.to_csv("gni_processed.csv", index=False)

print("Saved: gni_processed.csv")
```
