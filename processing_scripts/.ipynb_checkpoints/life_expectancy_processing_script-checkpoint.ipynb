```python
import pandas as pd

# Load dataset
df = pd.read_csv("life_expectancy.csv")

# Keep only years 2000–2023
df = df[(df["Year"] >= 2000) & (df["Year"] <= 2023)]


countries = [
    "United States", "Germany", "Japan", "Canada", "Australia",
    "China", "Brazil", "Mexico", "South Africa", "Turkey",
    "India", "Vietnam", "Zimbabwe", "Egypt", "Bangladesh",
    "Ethiopia", "Afghanistan", "Niger", "Chad", "Somalia",
    "World"
]

country_data = df[df["Entity"].isin(countries)]

# Round country/world values
country_data["Life expectancy"] = country_data["Life expectancy"].round(2)


categories = {
    "High income": ["United States", "Germany", "Japan", "Canada", "Australia"],
    "Upper middle income": ["China", "Brazil", "Mexico", "South Africa", "Turkey"],
    "Lower middle income": ["India", "Vietnam", "Zimbabwe", "Egypt", "Bangladesh"],
    "Low income": ["Ethiopia", "Afghanistan", "Niger", "Chad", "Somalia"]
}

category_rows = []

for category, country_list in categories.items():
    temp = df[df["Entity"].isin(country_list)]
    
    avg = (
        temp.groupby("Year")["Life expectancy"]
        .mean()
        .round(2)         
        .reset_index()
    )
    
    avg["Entity"] = category
    avg["Code"] = category.replace(" ", "_").lower()
    
    category_rows.append(avg)

category_data = pd.concat(category_rows)



final_df = pd.concat([country_data, category_data])

# Keep only needed columns
final_df = final_df[["Entity", "Code", "Year", "Life expectancy"]]

# Save new file
final_df.to_csv("life_expectancy.csv", index=False)

print("Saved:life_expectancy.csv")
```
